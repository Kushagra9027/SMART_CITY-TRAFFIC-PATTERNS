```python
import pandas as pd

df = pd.read_csv("../data/train_aWnotuB.csv")

df.head()
```

```python
df['DateTime'] = pd.to_datetime(df['DateTime'])
```

```python
df['year'] = df['DateTime'].dt.year
df['month'] = df['DateTime'].dt.month
df['day'] = df['DateTime'].dt.day
df['hour'] = df['DateTime'].dt.hour
df['weekday'] = df['DateTime'].dt.day_name()
```

```python
import matplotlib.pyplot as plt

df.groupby('Junction')['Vehicles'].mean().plot(kind='bar')

plt.title("Average Traffic by Junction")

plt.show()
```

```python
df.groupby('hour')['Vehicles'].mean().plot()

plt.title("Average Traffic by Hour")

plt.show()
```

```python
df['DateTime'] = pd.to_datetime(df['DateTime'])
df['weekday'] = df['DateTime'].dt.day_name()
df.groupby('weekday')['Vehicles'].mean().plot(kind='bar')
plt.title("Average Traffic by Weekday")
plt.xlabel("Weekday")
plt.ylabel("Average Vehicles")
plt.show()
```

```python
plt.figure(figsize=(15,5))

plt.plot(df['DateTime'],
         df['Vehicles'])

plt.title("Traffic Pattern Over Time")

plt.show()
```

```python
import seaborn as sns

sns.boxplot(
    x='Junction',
    y='Vehicles',
    data=df
)

plt.show()
```
